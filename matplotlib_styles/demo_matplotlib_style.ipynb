```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

## Using built-in styles ##

```python
print(plt.style.available)
```

```python
# plt.style.use('classic')
plt.style.use('seaborn-notebook')
plt.plot(x,y)
plt.title('Test Plot')
plt.xlabel("X")
plt.ylabel("sinX")
plt.legend(['sin'])
```

## User-defined style sheets ##

```python
plt.style.use('./publication_01.mplstyle')
# plt.style.use('./investigation_01.mplstyle')
plt.plot(x,y)
plt.title('Test Plot')
plt.xlabel("X")
plt.ylabel("sinX")
plt.legend(['sin'], loc="lower right")
```

## Dynamic Styling ##

```python
mpl.rc('figure', facecolor='white')
mpl.rc('lines', linewidth=4, marker='o', markersize=10)
mpl.rc('font', size=16, weight='normal')
mpl.rc('axes', linewidth=2, grid=True, labelsize='large', labelweight='bold')
mpl.rc('legend', frameon=True, fancybox=True, fontsize='large')
mpl.rc('xtick', labelsize = 'medium')
mpl.rc('ytick', labelsize = 'medium')
```
