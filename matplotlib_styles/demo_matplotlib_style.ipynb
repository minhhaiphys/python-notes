```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
def plot(sty):
    print("Use style: %s" %sty)
    plt.style.use(sty)
    plt.figure(figsize=(6,4))
    plt.plot(x,y,'.-',label=r"sin$\alpha$")
    plt.plot(x,z,'o--',label=r"cosin$\alpha$")
    plt.xlabel(r"$\alpha$ [rad]")
    plt.ylabel(r"Value $R_{MTJ}$")
    plt.title("style=%s" %sty)
    plt.legend()
    plt.tight_layout()
    plt.show()
    plt.close()
```

## Using built-in styles ##

```python
built_ins = plt.style.available
print(built_ins)
```

```python
%matplotlib inline
for sty in built_ins:
    plot(sty)
```

## User-defined style sheets ##

```python
plot('./clear.mplstyle')
```

```python
plot('./publish.mplstyle')
```

## Dynamic Styling ##

```python
mpl.rc('figure', facecolor='white')
mpl.rc('lines', linewidth=4, marker='o', markersize=10)
mpl.rc('font', size=16, weight='normal')
mpl.rc('axes', linewidth=2, grid=True, labelsize='large', labelweight='bold')
mpl.rc('legend', frameon=True, fancybox=True, fontsize='large')
mpl.rc('xtick', labelsize = 'medium')
mpl.rc('ytick', labelsize = 'medium')
```
